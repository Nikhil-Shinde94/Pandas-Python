# load the Data set

```
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data" 
iris_2 = pd.read_csv(url)
```

```
iris_2.head()
```

```
iris_2.columns = ['sepal_length','sepal_width', 'petal_length', 'petal_width', 'class']
iris_2.head()
```

**Is there any missing values**

```
iris_2.isnull().sum()
```

Step 6. Lets set the values of the rows 10 to 29 of the column 'petal_length' to NaN

```
iris_2.iloc[10:30, 2:3] = np.nan
iris_2.head(20)
```

step 7. Good, now lets substitute the NaN values to 1.0

```
iris_2.petal_length.fillna(1, inplace=True)
```

```
iris_2.head(20)
```

Step 8. Now let's delete the column class

```
del iris_2['class']
iris_2
```

Step 9. Set the first 3 rows as NaN

```
iris_2.iloc[0:3, :] = np.nan
iris_2.head()
```

Step 10. Delete the rows that have NaN

```
iris_2 = iris_2.dropna(how='any')
iris_2
```

Step 11. Reset the index so it begins with 0 again

```
iris_2 = iris_2.reset_index(drop=True)
iris_2.head()
```
