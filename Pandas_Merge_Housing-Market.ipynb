```
import pandas as pd
import numpy as np
```

```
#Create 3 differents Series, each of length 100, as follows:

num1 = pd.Series(np.random.randint(1, high=5, size=100, dtype='l'))
num2 = pd.Series(np.random.randint(1, high=4, size=100, dtype='l'))
num3 = pd.Series(np.random.randint(10000, high=30001, size=100, dtype='l'))

print(num1,num2,num3)
```

```
#Let's create a DataFrame by joinning the Series by column

housemarket = pd.concat([num1, num2, num3], axis=1)
housemarket.head()
```

```
#Change the name of the columns to bedrs, bathrs, price_sqr_meter

housemarket.rename(columns = {0: 'bedrs', 1: 'bathrs', 2: 'price_sqr_meter'}, inplace=True)
housemarket.head()
```

```
#Create a one column DataFrame with the values of the 3 Series and assign it to 'bigcolumn'

bigcolumn = pd.concat([num1, num2, num3], axis = 0)

bigcolumn = bigcolumn.to_frame()
```
